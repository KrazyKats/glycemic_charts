```python
import pandas as pd
from pathlib import Path
```

```python
demographics = pd.read_csv(Path('../data') / 'Demographics.csv')
demographics.head()
```

```python
glucose_spikes = pd.read_csv(Path('../website_data') / 'glucose_spikes.csv')
glucose_spikes.head()
```

```python
glucose_new = pd.merge(left=glucose_spikes, right=demographics, left_on="patient_id", right_on="ID")
glucose_new.head()
```

```python
glucose_new.to_csv("../website_data/glucose_spikes_patient.csv", index=False)
```
